```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gp
```

```python
events_filepath = '../data/raw/511_NY_Sporting__Concert__and_Special_Events__Beginning_2010.csv'

dataset = pd.read_csv(events_filepath)

dataset['Create Time'] = pd.to_datetime(dataset['Create Time'])
dataset.set_index('Create Time',inplace=True)

print(dataset.loc['2015-12-01': '2016-02-01'].count)
```

```python
dataset['Latitude'].hist(range=[40.6,40.9],bins=100)
```

```python
dataset['Longitude'].hist(range=[-74,-73],bins=100)
```

```python
# drop rows with lat/lng wrong (0,0)

df_map = dataset[dataset['Latitude'] > 40]
df_map = df_map[df_map['Longitude'] < -73]
```

```python
longitude = list(df_map.Longitude)
latitude = list(df_map.Latitude)
plt.figure(figsize = (10,10))
plt.plot(longitude,latitude,'.', alpha = 0.4, markersize = 0.5)
plt.show()

```
